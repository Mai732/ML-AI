```python
import pandas as pd
import numpy as np

# 1. Data Loading & Inspection
print("="*50)
print("STEP 1: Data Loading & Inspection")
print("="*50)

# Load the CSV file
df = pd.read_csv('company_data.csv')

# Print first 10 rows
print("\nFirst 10 rows of the dataset:")
print(df.head(10))

# Show data types and check for missing values
print("\nData Types and Missing Values:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# 2. Handling Missing Values
print("\n" + "="*50)
print("STEP 2: Handling Missing Values")
print("="*50)

# Fill missing Salary with median salary
salary_median = df['Salary'].median()
df['Salary'].fillna(salary_median, inplace=True)

# Fill missing Bonus with 0
df['Bonus'].fillna(0, inplace=True)

print(f"Missing Salary values filled with median salary: ${salary_median:.2f}")
print(f"Missing Bonus values filled with 0")
print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
# 3. Data Cleaning: Split Employee_Name
print("\n" + "="*50)
print("STEP 3: Splitting Employee Names")
print("="*50)

# Split Employee_Name into First_Name and Last_Name
df[['First_Name', 'Last_Name']] = df['Employee_Name'].str.split(' ', expand=True)

# Delete the original Employee_Name column
df.drop('Employee_Name', axis=1, inplace=True)

print("First 5 rows after splitting names:")
print(df[['First_Name', 'Last_Name', 'Department']].head())
```

```python
# 4. Filtering: Create Senior_IT DataFrame
print("\n" + "="*50)
print("STEP 4: Creating Senior_IT DataFrame")
print("="*50)

Senior_IT = df[(df['Department'] == 'IT') & (df['Years_Experience'] > 8)]

print(f"Number of Senior IT employees: {len(Senior_IT)}")
print("\nSenior IT employees:")
print(Senior_IT[['First_Name', 'Last_Name', 'Years_Experience']])
```

```python
# 5. Performance Mapping
print("\n" + "="*50)
print("STEP 5: Creating Performance_Level column")
print("="*50)

def assign_performance_level(score):
    if score > 80:
        return 'High'
    elif score >= 50:
        return 'Medium'
    else:
        return 'Low'

df['Performance_Level'] = df['Evaluation_Score'].apply(assign_performance_level)

print("Performance Level distribution:")
print(df['Performance_Level'].value_counts())
print("\nFirst 5 rows with Performance_Level:")
print(df[['First_Name', 'Evaluation_Score', 'Performance_Level']].head())
```

```python
# 6. Calculate total salary expense and average evaluation score by Department
print("\n" + "="*50)
print("STEP 6: Department-wise Analysis")
print("="*50)

dept_analysis = df.groupby('Department').agg({
    'Salary': 'sum',
    'Evaluation_Score': 'mean'
}).round(2)

dept_analysis.columns = ['Total_Salary_Expense', 'Average_Evaluation_Score']
print("Department-wise Analysis:")
print(dept_analysis)
```

```python
# 7. Find top 5 highest-paid employees
print("\n" + "="*50)
print("STEP 7: Top 5 Highest-Paid Employees")
print("="*50)

top_5_paid = df.nlargest(5, 'Salary')[['First_Name', 'Last_Name', 'Salary']]
print("Top 5 highest-paid employees:")
print(top_5_paid)
```

```python
# 8. Give 10% raise to Sales department
print("\n" + "="*50)
print("STEP 8: 10% Raise for Sales Department")
print("="*50)

# Before raise
sales_before = df[df['Department'] == 'Sales']['Salary'].mean()
print(f"Average Sales salary before raise: ${sales_before:.2f}")

# Apply 10% raise
df.loc[df['Department'] == 'Sales', 'Salary'] *= 1.10

# After raise
sales_after = df[df['Department'] == 'Sales']['Salary'].mean()
print(f"Average Sales salary after raise: ${sales_after:.2f}")
```

```python
# 9. Exporting Results
print("\n" + "="*50)
print("STEP 9: Exporting Final Report")
print("="*50)

# Save to CSV
df.to_csv('final_employee_report.csv', index=False)

# Also save to Excel for demonstration
df.to_excel('final_employee_report.xlsx', index=False)

print("Files saved successfully:")
print("- final_employee_report.csv")
print("- final_employee_report.xlsx")

# Display final DataFrame info
print("\nFinal DataFrame Info:")
print(df.info())
print("\nFirst 5 rows of final DataFrame:")
print(df.head())
```
